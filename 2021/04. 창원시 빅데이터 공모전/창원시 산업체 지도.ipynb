## 겹치는 부분 확대해서 보기
+ 덕산일반산업지구
+ 마산자유무역지구
+ 평성일반산업단지
+ 창원국가산업단지
+ 한국산업단지공단창원



```python
import folium
#지도 중심 만들기
lat = store['Latitude'].mean()
long = store['Longitude'].mean()
s = folium.Map([lat,long],zoom_start=12)

for i in store.index:
    sub_lat =  store.loc[i,'Latitude']
    sub_long = store.loc[i,'Longitude']
    
    title = store.loc[i,'도로명주소']
    color = 'navy' #'#352587'
    #지도에 동그라미로 데이터 찍기    
    folium.CircleMarker([sub_lat,sub_long],color=color,radius = 500, fill_color='#7678B1',tooltip=title).add_to(s)

for i in fac.index:
    sub_lat =  fac.loc[i,'Latitude']
    sub_long = fac.loc[i,'Longitude']
    
    title = fac.loc[i,'도로명주소']
    color = 'cyan'
    #지도에 동그라미로 데이터 찍기    
    folium.CircleMarker([sub_lat,sub_long],color=color,radius = 500, tooltip=title).add_to(s)

s
```

+ 마산자유무역지구 : 산업단지와 상권이 겹친다.

```python
store1 = store
store1['아파트'] = store1['도로명주소'].str[-3:]
store1['아파트']
```

```python
a = store1[store1['아파트']== '아파트'].index
store1 = store1.drop(a)
store1 #36개

b = store1[store1['아파트']== '등학교'].index
store1 = store1.drop(b)
store1 #11개

store1 = store1.drop(['아파트'],axis=1)
store1
```
