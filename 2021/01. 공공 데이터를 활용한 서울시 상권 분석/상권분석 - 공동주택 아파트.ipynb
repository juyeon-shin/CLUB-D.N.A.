## 2. 구별로 분할

```python
#본인 구 설정
dj = apartment['주소(시군구)'] == '동작구'
dj = apartment[dj]

gw = apartment['주소(시군구)'] == '관악구'
gw = apartment[gw]

gch = apartment['주소(시군구)'] == '금천구'
gch = apartment[gch]
```

### 2.1 해당 구의 동을 기준으로 갯수 세기

```python
#한 번 해봤는데 큰 의미는 없음...
dj.groupby(['주소(읍면동)']).count()
gw.groupby(['주소(읍면동)']).count()
gch.groupby(['주소(읍면동)']).count()
```

### 2.2 동 기준, 합 보기 (전체 세대 수 합), 그래프 그리기

```python
dj_DONG = dj.groupby(['주소(읍면동)'], as_index = False).sum()
gw_DONG = gw.groupby(['주소(읍면동)'],as_index = False).sum()
gch_DONG = gch.groupby(['주소(읍면동)'],as_index = False).sum()

#구 전체 합
dj_DONG['k-전체세대수'].sum()
gw_DONG['k-전체세대수'].sum()
gch_DONG['k-전체세대수'].sum()

#구 전체 합 데이터 프레임 만들기
GU_count = pd.DataFrame({'구' : ['동작구','관악구','금천구'],
                        '전체 세대수 합' : [dj_DONG['k-전체세대수'].sum(),gw_DONG['k-전체세대수'].sum(),gch_DONG['k-전체세대수'].sum()]})

GU_count
```

```python
# 구의 전체 세대수 합 그래프 그리기
GU_count_ax = sns.barplot(x = '구', y='전체 세대수 합',data = GU_count,palette="Blues_d")
```

```python
# 각 구의 동 기준으로 전체 세대수 그래프 그리기
dj_ax = sns.barplot(x='주소(읍면동)', y="k-전체세대수", data=dj_DONG, orient= 'v', palette="Blues_d")
```

동작구는 상도동의 아파트 세대수가 제일 많다
동작동이 아파트 세대 수가 제일 없다

```python
gw_ax = sns.barplot(x='주소(읍면동)', y="k-전체세대수", data=gw_DONG, orient= 'v',palette="Blues_d")
```

관악구는 봉천동의 아파트 세대수가 제일 많다
남현동이 아파트 세대 수가 제일 없다
실제로 남현동은 출입금지구역이 많음
-> 남현동의 상권이 좋지 않을 것으로 예상

```python
gch_ax = sns.barplot(x='주소(읍면동)', y="k-전체세대수", data=gch_DONG, orient= 'v',palette="Blues_d")
```

금천구는 시흥동 아파트 세대 수가 제일 많다

가산동은 서울 디지털 산업단지가 있어서 아파트 자체가 적은듯
-> 가산동의 상권을 좋지 않을 것으로 예상
