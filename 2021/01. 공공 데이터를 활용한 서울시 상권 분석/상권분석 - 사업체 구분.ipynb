## 상권분석 - 사업체구분
### 1. 데이터 불러오기

```python
#모듈 불러오기
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc("font",family='Malgun Gothic')
```

```python
df1=pd.read_csv('C:/Users/juyeo/Downloads/행정구역별_사업체_현황.csv')

df1.columns=['자치구','산업대분류','사업체_1','사업체_2','사업체_3','사업체_4','항목','단위','2018년']
df1.head(5)
```

```python
df2= df1[df1['산업대분류'] == '합계'].index
df3=df1.drop(df2)
df3.head(3)
```

```python
df3['산업대분류'].value_counts()
plt.figure(figsize=[20,5])

sns.countplot(data=df3, y='산업대분류')
plt.legend(loc='right')
```

```python
#관악, 금천, 동작의 사업체 분류

df2=df1[df1['산업대분류']=='합계'].index
df3=df1.drop(df2)
sns.barplot(data=df3, x='자치구',order=['관악구','금천구','동작구'],y='2018년',hue='산업대분류')
plt.figure(figsize=[40,5])
```
