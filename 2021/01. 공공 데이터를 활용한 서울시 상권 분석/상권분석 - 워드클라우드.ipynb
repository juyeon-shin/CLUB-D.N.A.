### 4. 빈도 그래프 만들기
빈도순으로 단어를 정렬해서 막대그래프를 만드는 과정
워드클라우드 만들 때 빈도수가 포함된 사전을 가지고 만들어줄거라 막대그래프는 필수는 아님! 빈도수 사전이 중요해

```python
# 단어 빈도수 세기

def counting(string):
    count = {}
    for word in string:
        count[word] = count.get(word, 0) + 1
    return count
```

```python
count_kj = counting(kjg)
count_jl = counting(jlg)
count_kb = counting(kbg)
```

```python
# 막대그래프 만들기

import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

def word_graph(cnt, max_words=10):
    
    sorted_w = sorted(cnt.items(), key=lambda kv: kv[1])
    # print(sorted_w[-max_words:])
    n, w = zip(*sorted_w[-max_words:])

    plt.barh(range(len(n)),w,tick_label=n)
    plt.show()
```

```python
word_graph(count_kj, max_words=20)
```

### 5. 워드 클라우드 만들기
앞에서 토큰화된 결과를 가지고 빈도수를 센 사전으로 워드클라우드를 만드는거야
원모양으로 하고싶어서 찾다가 그냥 동전모양 그림 구글에서 다운받아서 그 위에 워드클라우드 그리는걸로 했는데
먼저 이미지파일을 컴퓨터에 저장해줘야되고!
함수 안에 만들어진 워드클라우드를 저장하는 경로도 각자 컴퓨터에 맞춰서 바꿔주기
저장하는 파일 이름이 같으면 계속 덮어씌워질 수 있으니까 다른 구역으로 실행할 땐 이름 바꿔주는게
좋을 것 같아

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image
# generate() 를 사용한 원문에서의 자동생성

def createwc(cnt):
    
    coin_mask = np.array(Image.open("C:/Users/USER/Desktop/동전.jpg")) # 이미지 저장하고 여기 주소 바꾸기!
    
    font_path = 'c:/Windows/Fonts/malgun.ttf'

    wc = WordCloud(background_color="white",
               font_path = font_path,# 배경색 지정
               max_words=50, # 출력할 최대 단어 수
               width = 1000, #이미지 너비 지정
               height = 600, #이미지 높이 지정
               mask=coin_mask, # 배경으로 사용할 이미지
               contour_width=5,  # 테두리선의 크기
               contour_color='white') # 테두리선의 색
               #background_color='white', #이미지 배경색 지정
    
    wc.generate_from_frequencies(cnt)
    
    plt.figure()
    plt.axis("off")
    plt.imshow(wc, interpolation="bilinear")
    wc.to_file("C:/Users/USER/Desktop/wordcloud.png")
    # 컴퓨터에 워드클라우드 저장할 경로!! 두개 만들면 겹치니까 만들때마다 이름 바꿔주면 좋을듯
    plt.show()
```
