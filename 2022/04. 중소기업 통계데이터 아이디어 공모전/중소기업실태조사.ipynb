```python
#원하는 분야 선택

def extract_frame(col_name):
    col = pd.DataFrame()
    for i in md_ser.columns:
        if col_name in i:
            col= pd.concat([col,md_ser[i]],axis=1)
    return col
```

### 02. 위탁기업과 거래시 주로 겪은 애로사항 & 지원 사항

+ 15-32 : 위탁기업과 거래시 주로 겪은 애로사항
+ 32-47 : 위탁기업으로부터 지원 받은 사항

```python
def select_columns(num1,num2):
    GEN_F_select = GEN_F.iloc[:,num1:num2]
    GEN_F_select

    select = pd.DataFrame()
    name_list = []
    for i in GEN_F_select.columns:
        f_s = GEN_F_select.groupby([i]).count().iloc[:,0]
        name_list.append(f_s.index.name)
        select = pd.concat([select,f_s],axis=1)

    select.columns = name_list
    return select
```

```python
def select_plot(select_f,f_title):
    plt.figure(figsize = (10,6))
    sns.barplot(x=select_f.columns,y=select_f.values[1])
    plt.xticks(rotation=45)
    plt.title(f_title)
    plt.show()
```

```python
select_f = select_columns(15,32)
select_plot(select_f,'위탁기업과 거래시 주로 겪은 애로사항')
```

+ GEN_F_32 : 납기 단축, 촉박
+ GEN_F_25 : 원자재가격상승분 등 납품단가 미반영
+ GEN_F_33 : 납품대금 결제기한 미준수

```python
select_f = select_columns(32,47)
select_plot(select_f,'위탁기업으로부터 지원 받은 사항')
```

+ GEN_F_48 : 원자재 구매 지원
