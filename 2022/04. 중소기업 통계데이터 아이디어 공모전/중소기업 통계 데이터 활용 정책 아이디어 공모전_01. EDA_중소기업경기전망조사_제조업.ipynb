```python
def df_complain(data):
    df = data.groupby(['date']).count()
    df = df[['X55','X56', 'X57', 'X58', 'X59', 'X60', 'X61', 'X62', 'X63', 'X64', 'X65','X66', 'X67', 'X68', 'X69', 'X70', 'X71']]
    
    index_list = []
    for i in df.index:
        index_list.append(i.split('.')[0])

    df.index = index_list
    return df
```

```python
for i in mnX3:
    read_dt = pd.read_csv(i)
    read_dt = df_complain(read_dt)
    
read_dt
```
