```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('/content/drive/Shareddrives/202206 나의 분석일지/다이아몬드 가격 예측/Diamonds Prices2022.csv')
```

```python
train = train.drop(['Unnamed: 0'],axis=1)
train
```

+ 수치형 변수 : carat, x,y,z
+ 범주형 변수 : color, color/clarity

color/clarity 는 파생변수

```python
train_c = train.copy()
train_c['color/clarity'] = train_c['color']+'/'+train_c['clarity']
train_c = train_c.drop(['cut','clarity','depth','table'],axis=1)
train_c
```

```python
X = train_c.drop(['price'],axis=1)
y = train_c['price']
```

```python
#import catboost as cb
#import shap 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.inspection import permutation_importance
```

```python
#!pip install catboost
import catboost as cb
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=2022)
```

```python
train_data = cb.Pool(X_train, y_train)
test_data = cb.pool(X_test, y_test)
```

```python
model = cb.CatBoostRegressor()
```

```python
grid = {'iterations': [100, 150, 200],
        'learning_rate': [0.03, 0.1],
        'depth': [2, 4, 6, 8],
        'l2_leaf_reg': [0.2, 0.5, 1, 3]}
model.grid_search(grid, train_data)
```
