### 01. at파일에서 농산물 추출

```python
train_list= glob.glob('C:/Users/juyeo/Desktop/03. 농산물 가격예측 AI 경진대회/public_data/train_AT_TSALET_ALL/*.csv')
```

```python
def extraction(product):
    train_at_feature = pd.DataFrame()

    for train in train_list:
        current_data = pd.read_csv(train)
        for i in product:
            new_data = current_data[current_data['PUM_NM'] == i]
            train_at_feature = pd.concat([train_at_feature,new_data])

    return train_at_feature
```

```python
product = ['배추','양파','건고추','새송이']
train_4 = extraction(product)
```

```python
def ext_algi(train_4,algi):
    train_4_c= train_4.drop(['WHSAL_NM','CMP_NM','DAN_NM','POJ_NM','SIZE_NM','LV_NM'],axis=1)
    train_1 = train_4_c[train_4_c['PUM_NM'] == algi]
    train_1.reset_index(inplace=True,drop=True)
    return train_1
```

```python
algi_1 = ext_algi(train_4,'배추')
algi_2 = ext_algi(train_4,'양파')
algi_3 = ext_algi(train_4,'새송이')
algi_4 = ext_algi(train_4,'건고추')
```

### 02. 날짜 및 지역 전처리
#### 02-1. 날짜

```python
def pret_date(data):
    train_date = data['SALEDATE'].astype('str')
    
    for i in range(len(train_date)):
        if len(train_date[i]) == 8:
            train_date[i] = train_date[i][:4]+'-'+train_date[i][4:6]+'-'+train_date[i][6:]
        else :
            pass
        
    return train_date
```

#### 02-2. 지역 전처리

```python
def pret_addr(data):
    SAN_NM = data['SAN_NM'].astype('str')
    split_1 = SAN_NM.str.split(' ')
    place = split_1.str[0]
    
    for i in range(len(place)):
        if '광역시' in place[i]:
            place[i] = place[i][:-3]
        else :
            pass
        
    for i in range(len(place)):
        if place[i] == '서울':
            place[i] = '서울경기'
        elif place[i] == '서울특별시':
            place[i] = '서울경기'
        elif place[i] == '인천':
            place[i] = '서울경기'
        elif place[i] == '경기':
            place[i] = '서울경기'
        elif place[i] == '경기도':
            place[i] = '서울경기'
        elif place[i] == '강원도':
            place[i] = '강원'
        elif place[i] == '충청남도':
            place[i] = '충남'
        elif place[i] == '대전':
            place[i] = '충남'
        elif place[i] == '충청북도':
            place[i] = '충북'
        elif place[i] == '세종시':
            place[i] = '충북'
        elif place[i] == '전라남도':
            place[i] = '전남'
        elif place[i] == '광주':
            place[i] = '전남'
        elif place[i] == '전라북도':
            place[i] = '전북'
        elif place[i] == '경상남도':
            place[i] = '경남'
        elif place[i] == '부산':
            place[i] = '경남'
        elif place[i] == '경상북도':
            place[i] = '경북'
        elif place[i] == '울산':
            place[i] = '경북'
        elif place[i] == '대구':
            place[i] = '경북'
        elif place[i] == '제주도':
            place[i] = '제주'
        else :
            place[i] = '전국'
    
    return place
```

```python
def make_data(data):
    make_data = pd.DataFrame()
    make_data['날짜'] = pret_date(data)
    make_data['지점'] = pret_addr(data)
    return make_data
```

```python
data_cabbage = make_data(algi_1)
data_onion = make_data(algi_2)
data_oyster = make_data(algi_3)
data_redpepper = make_data(algi_4)
```

```python
train_temp= glob.glob('C:/Users/juyeo/Desktop/03. 농산물 가격예측 AI 경진대회/temp_rain/*_기온.csv')
train_rain= glob.glob('C:/Users/juyeo/Desktop/03. 농산물 가격예측 AI 경진대회/temp_rain/*_강수.csv')
```

```python
def merge_train(argi,weather):
    new = pd.DataFrame()

    for i in argi['지점'].unique():
        for train in weather:
            if i == train[-9:-7] or i == train[-11:-7]:
                unit = pd.read_csv(train,skiprows=7, encoding='cp949')
                con = pd.merge(argi[argi['지점']==i],unit,on=['날짜','지점'],how='right')
                new = pd.concat([new,con],axis=0)

    return new
```

```python
def make_argi(argi):
    temp = merge_train(argi,train_temp)
    rain = merge_train(argi,train_rain)
    temp['강수량(mm)'] = rain['강수량(mm)']
    data = temp.groupby(['날짜']).mean()
    data.index = pd.to_datetime(data.index)
    return data
```

```python
cabbage = make_argi(data_cabbage)
onion = make_argi(data_onion)
oyster = make_argi(data_oyster)
redpepper = make_argi(data_redpepper)
```

### 04. 시계열 분해

```python
Train = pd.read_csv('public_data/train.csv')
Train
```

```python
def sep_timeseries(Train, algi):
    from scipy import stats
    import statsmodels.api as sm
    from statsmodels.tsa.seasonal import seasonal_decompose
    
    Train.index = pd.to_datetime(Train['date'])
    Train.set_index(Train['date'])
    result = seasonal_decompose(Train[algi],model ='additive')
    #result.plot()
    
    re = sm.tsa.seasonal_decompose(Train[algi],model ='additive')
    
    new = pd.DataFrame()
    new['가격'] = re.observed
    new['trand'] = re.trend
    new['seasonal'] = re.seasonal
    new['resid'] = re.resid

    return new
```

```python
sep_timeseries(Train,'배추_가격(원/kg)')
```

### 05. 데이터셋 재구성

```python
cabbage = pd.concat([Train['배추_거래량(kg)'],sep_timeseries(Train,'배추_가격(원/kg)'),cabbage],axis=1)
cabbage.head()
```

```python
onion = pd.concat([Train['양파_거래량(kg)'],sep_timeseries(Train,'양파_가격(원/kg)'),onion],axis=1)
onion.head()
```

```python
oyster = pd.concat([Train['새송이_거래량(kg)'],sep_timeseries(Train,'새송이_가격(원/kg)'),oyster],axis=1)
oyster.head()
```

```python
redpepper = pd.concat([Train['건고추_거래량(kg)'],sep_timeseries(Train,'건고추_가격(원/kg)'),redpepper],axis=1)
redpepper.head()
```
